```python
import requests, pathlib, hashlib

def download_video_file(url):
    dst = pathlib.Path("ads")                           
    dst.mkdir(exist_ok=True)

    fname = hashlib.md5(url.encode()).hexdigest() + ".mp4"
    path  = dst / fname

    with requests.get(url, stream=True, timeout=30) as r:
        r.raise_for_status()                   
        with path.open("wb") as f:
            for chunk in r.iter_content(2**16):
                f.write(chunk)

    print("saved to", path, '\n')
    return path
```

```python
snapshot_videos = ads['snapshot.videos']

for i, links in enumerate(snapshot_videos):
    print(f"ADS {i+1}")

    try:
        links = links[0]
    except:
        print("No video available\n")   
        continue 
    
    for key in links:
        print(f"- {key}: {links[key]}")
        if key == "video_sd_url":
            download_video_file(links[key])
```
